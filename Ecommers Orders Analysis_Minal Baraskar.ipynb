```python
#17.Plot a line chart showing sales (TotalAmount) over Ɵme (OrderDate).
```

```python
import matplotlib.pyplot as plt
```

```python
df = df.sort_values('OrderDate')
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['OrderDate'],df['TotalAmount'],color = "Blue",marker = "o")
plt.xlabel("Order Date")
plt.ylabel("Total Amount")
plt.title("Sales Over Time")
```

```python
#18.Plot a bar chart showing total sales by region.
```

```python
#19.Create a pie chart for product sales distribuƟon.
```
